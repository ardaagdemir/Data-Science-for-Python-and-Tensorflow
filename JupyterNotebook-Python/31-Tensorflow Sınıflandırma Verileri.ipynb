```python
#x ve y leri ayarlayıp 2 ye bölelim

y=dataFrame["Type"].values
x=dataFrame.drop("Type", axis=1).values
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.3, random_state=15)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
scaler.fit(x_train)
```

```python
x_train = scaler.transform(x_train)
```

```python
x_test = scaler.transform(x_test)
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation, Dropout
from tensorflow.keras.callbacks import EarlyStopping
```

```python
#EarlyStopping ve Dropout özellikle over feeding problemlerini çözmek için kullanacağız
```

```python
x_train.shape
```

```python
model = Sequential()

model.add(Dense(units = 30, activation="relu")) #kaç kolon var ise giriş layerının(units) o sayıda olması önerilir
model.add(Dense(units = 15, activation="relu"))
model.add(Dense(units = 15, activation="relu"))
model.add(Dense(units = 1, activation="sigmoid")) #sigmoid 0 ve 1 arasında değer verir ve çıkış layerına koyulabilir
#1 layerdan fazla durumlarda 30 ve 1 arasında sayılar olması yine önerilen bir durumdur

model.compile(loss="binary_crossentropy", optimizer = "adam") 
#sınıflandırma yaptığüımız için loss değeriini önceki örneklere göre farklı seçtik
```

```python
model.fit(x=x_train, y=y_train, epochs=700, validation_data=(x_test, y_test), verbose=1) 
#over feeding için epochs u yüksek bir değer seçtik
```

```python
modelKaybi = pd.DataFrame(model.history.history)
```

```python
modelKaybi.plot()
```

```python
#Görüldüğü gibi önceki örneklerden çok daha farklı bir grafik oldu.
#Bunun nedeni epochs un yüksek bir değer girilmesi yani over feeding oldu
#Bu sapmanın önüne geçmek için epochs u bir yerde durmamız gerekiyor.
#Validation loss sapma miktarı arttığında epochs durmalıdır. Bunu yapmak için Earlystopping denilen mekanizmayı kullanmalıyız
```

```python
model = Sequential()

model.add(Dense(units = 30, activation="relu")) 
model.add(Dense(units = 15, activation="relu"))
model.add(Dense(units = 15, activation="relu"))
model.add(Dense(units = 1, activation="sigmoid")) 

model.compile(loss="binary_crossentropy", optimizer = "adam") 
```

```python
earlyStopping = EarlyStopping(monitor="val_loss", mode="min", verbose=1, patience=25)
```

```python
model.fit(x=x_train, y=y_train, epochs=700, validation_data=(x_test, y_test), verbose=1, callbacks=[earlyStopping])
```

```python
modelKaybi= pd.DataFrame(model.history.history)
```

```python
modelKaybi.plot()
```

```python
#sonuçtan hala memnun değil isek burada ekstra oalrak "Dropout" kullanabiliriz
#"Dropout", verdiğimiz rate kaç ise o kadar layer ı modelimizden rastgele atmaya başlayacak ve daha iyi bir sonuç elde edeceğiz
```

```python
model = Sequential()

model.add(Dense(units = 30, activation="relu"))
model.add(Dropout(0.6))

model.add(Dense(units = 15, activation="relu"))
model.add(Dropout(0.6))

model.add(Dense(units = 15, activation="relu"))
model.add(Dropout(0.6))

model.add(Dense(units = 1, activation="sigmoid")) 

model.compile(loss="binary_crossentropy", optimizer = "adam") 
```

```python
model.fit(x=x_train, y=y_train, epochs=700, validation_data=(x_test, y_test), verbose=1, callbacks=[earlyStopping])
```

```python
kayipDf = pd.DataFrame(model.history.history)
```

```python
kayipDf.plot()
```

```python
#Görüldüğü gibi over feeding oldukça azalmış gözüküyor.
```

```python
#Sınıfları tahminet
tahminlerimiz = model.predict_classes(x_test)
```

```python
tahminlerimiz
```

```python
#sınıflandırmanun ne kadar düzgün sonuçlar verdiğini görmek için kullanılan methodlardır
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(classification_report(y_test, tahminlerimiz))
```

```python
print(confusion_matrix(y_test, tahminlerimiz))
```
