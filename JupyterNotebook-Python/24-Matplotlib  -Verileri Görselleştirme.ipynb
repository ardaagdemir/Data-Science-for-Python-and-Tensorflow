```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
yasListesi = [10,20,30,30,30,40,50,60,60,70]
```

```python
kiloListesi = [50,60,60,60,70,75,80,85,88,90]
```

```python
numpyYasListesi= np.array(yasListesi)
numpyKiloListesi= np.array(kiloListesi)
```

```python
numpyKiloListesi
```

```python
numpyYasListesi
```

```python
#
```

```python
#grafik x ve y eksenlerine listeleri yerleştirmek için
#yazılan g değeri green anlamına gelmektedir ve line colour ifade eder
plt.plot(numpyYasListesi,numpyKiloListesi, "g") 
#grafik eksenlerinin isimlerini belirlemek için
plt.xlabel("Yas")
plt.ylabel("Kilo")
#grafiğin başlığını belirlemek için
plt.title("Kilo' nun Yaşa Göre Değişim Grafiği")
```

```python
numpyDizisi1 = np.linspace(0,10,20)
```

```python
numpyDizisi1
```

```python
numpyDizisi2 = numpyDizisi1 ** 3
```

```python
numpyDizisi2
```

```python
plt.plot(numpyDizisi1,numpyDizisi2, "b--") #line gösterim şeklini değiştirebiliriz
```

```python
plt.plot(numpyDizisi1,numpyDizisi2, "b*-") #line gösterim şeklini değiştirebiliriz
```

```python
#grafikleri yan yana koyarak eksenlerde farklı verileri göstermek için aşağıdaki komutları uygulamamız gerekiyor
plt.subplot(1,2,1) #parantez içi şu anlama gelmektedir..  1 sıra, 2 kolon, 1. grafik
plt.plot(numpyDizisi1,numpyDizisi2, "r*-")
plt.subplot(1,2,2) #parantez içi şu anlama gelmektedir..  1 sıra, 2 kolon, 2.grafik
plt.plot(numpyDizisi2,numpyDizisi1)
```

```python
## Şimdiye kadar hazır kütüphaneleri kullanarak grafikler çizdirdik.
#Artık kütüphaneden bir obje oluşturarak oluşturulan obje üzerinde işlem yapacağız
```

```python
benimFigure = plt.figure()
figureAxes = benimFigure.add_axes([0.2,0.2,0.9,0.9]) ##grafiğin axes değerlerini bir liste olarak girmemiz gerekiyor
                                                     #son iki değer grafiğin yüksekliğini ve genişliğini ifade etmektedir
                                                     #ilk iki değer diğer not defterinde iç içe grafiklerde anlatılmakadır
figureAxes.plot(numpyDizisi1,numpyDizisi2, "g--")
figureAxes.set_xlabel("X ekseni")
figureAxes.set_ylabel("Y ekseni")
figureAxes.set_title("Grafik Başlığı")
```
