```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
numpyDizisi1 = np.linspace(0,10,20)
```

```python
numpyDizisi2 = numpyDizisi1 ** 3
```

```python
##grafiğin başlangıç değerlerinin nasıl bir fark yarattığını anlayacağız
figure2 = plt.figure()
eksen1 = figure2.add_axes([0.1,0.1,0.7,0.7]) #başlangıç değerleri(ilk iki değer) iki grafik içinde görüldüğü gibidir
eksen2 = figure2.add_axes([0.3,0.3,0.3,0.3])
eksen1.plot(numpyDizisi1,numpyDizisi2, "g")
eksen1.set_xlabel("X ekseni")
eksen1.set_ylabel("Y ekseni")
eksen1.set_title("Ana Grafik Başlık")

eksen2.plot(numpyDizisi2,numpyDizisi1, "b")
eksen2.set_xlabel("X ekseni")
eksen2.set_ylabel("Y ekseni")
eksen2.set_title("Küçük Grafik Başlık")
```

```python
## başlangıç değerlerini değiştirerek tekrar çizdirelim
```

```python
figure2 = plt.figure()

eksen1 = figure2.add_axes([0.1,0.1,0.7,0.7])

eksen2 = figure2.add_axes([0.2,0.4,0.3,0.3])

eksen1.plot(numpyDizisi1,numpyDizisi2, "g")
eksen1.set_xlabel("X ekseni")
eksen1.set_ylabel("Y ekseni")
eksen1.set_title("Ana Grafik Başlık")

eksen2.plot(numpyDizisi2,numpyDizisi1, "b")
eksen2.set_xlabel("X ekseni")
eksen2.set_ylabel("Y ekseni")
eksen2.set_title("Küçük Grafik Başlık")
```
