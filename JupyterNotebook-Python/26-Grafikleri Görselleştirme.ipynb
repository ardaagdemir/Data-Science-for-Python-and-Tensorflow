```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
numpyDizisi1 = np.linspace(0,20,10)
numpyDizisi2 = numpyDizisi1 ** 2
```

```python
yeniFigur = plt.figure()
##eksen boyutlarını girmeden figure ile de grafik boyutunu belirleyebiliriz (figsze=(3,3), figsize(5,4) gibi..)
##grafik kalitesi için figure kısmında (dpi=50, dpi=150 gibi..)
yeniEksen = yeniFigur.add_axes([0.1,0.1,0.9,0.9])

yeniEksen.plot(numpyDizisi1,numpyDizisi1 ** 2, label = "numpyDizisi ** 2")
yeniEksen.plot(numpyDizisi1,numpyDizisi1 ** 3, label = "numpyDizisi ** 3")
yeniEksen.legend() ##yerini belirlemek için hazır locasyonları deneyebiliriz (loc=1, loc=2, loc=3 gibi tanımlanmalıdır)
## label değeri grafik çizgisinin legend ını oluşturmamıza yaramaktadır
```

```python
##Grafiği kayıt etmek için;
yeniFigur.savefig("benimfigur.png", dpi=200)
```

## Görselleştirme de ileri seviye işlemler

```python
(benimFigure, benimEksen) = plt.subplots()
benimEksen.plot(numpyDizisi1, numpyDizisi2, "r")
##renk kodlarını bulabilmek için google da "rgb hex code" diye aratabiliriz
##bulunan kodu color="kod" şeklinde kod satırına yazabiliriz
benimEksen.plot(numpyDizisi2, numpyDizisi2, color="#C96F23")
```

```python
## grafik çizgilerinin kalınlıklarını ve stillerini değiştirebiliriz
(yeniFigur, yeniEksen) = plt.subplots()
yeniEksen.plot(numpyDizisi1,numpyDizisi2 + 2, color="yellow", linewidth = 5.0) ##linewidth kalınlığı değiştirir
yeniEksen.plot(numpyDizisi1 + 4, numpyDizisi2 + 5, color="red")
```

```python
(yeniFigur, yeniEksen) = plt.subplots()

yeniEksen.plot(numpyDizisi1 + 4, numpyDizisi2 + 5, color="red", linestyle="-.")

yeniEksen.plot(numpyDizisi1*2, numpyDizisi2+3, color="gray", linestyle=":") #diğer stiller ("--", "-.", "*-", ":")

##marker = çizgilerin eksenler ile kesiştiği noktaları belirtmek için kullanılır
yeniEksen.plot(numpyDizisi1*3, numpyDizisi2+3, color="black", linestyle="--", marker="+", markersize=10)
```
